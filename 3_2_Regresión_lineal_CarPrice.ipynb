 # Funci√≥n para seleccionar columnas con una correlaci√≥n espec√≠fica

```python
def correlation(dataframe,threshold):
  cols_corr = set()
  corr_matrix = dataframe.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i,j]) >= threshold:
        col_name = corr_matrix.columns[i]
        cols_corr.add(col_name)
  return cols_corr
```

```python
corr_cols = correlation(dataframe.iloc[:,:-1],0.40)
corr_cols
```

# Separar columnas categ√≥ricas

```python
def categorical(dataframe):
    categorical_col = []
    for col in dataframe:
     if dataframe[col].dtype == 'object':
        categorical_col.append(col)
    return categorical_col
```

```python
categorical_col = categorical(dataframe)
print(categorical_col)
```

# Obtener columnas numericas

```python
def numerical(dataframe):
  numerical_col = []
  for col in dataframe:
    if dataframe[col].dtype != 'object':
      numerical_col.append(col)
  return numerical_col
```

```python
numerical_col = numerical(dataframe)
print(numerical_col)
```

# Normalizar con **Min-Max**

```python
def minmax_norm(dataframe):
  return (dataframe - dataframe.min()) / (dataframe.max() - dataframe.min())
```

```python
print(dataframe['price'])
```

```python
temporal = minmax_norm(dataframe['price'])
print(temporal)
```
